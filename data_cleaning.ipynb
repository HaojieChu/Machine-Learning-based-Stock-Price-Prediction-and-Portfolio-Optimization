```python
import os
import pandas as pd

destination_directory_15y = '/Users/haojiechu/Documents/Python_documents/python_excercises/stock_data_15y_cleaned'
output_directory = '/Users/haojiechu/Documents/Python_documents/python_excercises/stock_data_15y_processed'

# Ensure the destination directory exists
if os.path.exists(destination_directory_15y):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # List all files in the destination directory
    files_in_15y = [file for file in os.listdir(destination_directory_15y) if os.path.isfile(os.path.join(destination_directory_15y, file))]

    # Loop through each file
    for file_name in files_in_15y:
        file_path = os.path.join(destination_directory_15y, file_name)

        # Check if the file is a CSV file (you can modify the condition for other file types)
        if file_name.endswith('.csv'):
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)

            # Check if the DataFrame has at least one row
            if not df.empty:
                # Calculate the 'H-L' column (High - Low)
                df['H-L'] = df['High'] - df['Low']

                # Calculate the '7 DAYS MA' column (Close seven days moving average)
                df['7 DAYS MA'] = df['Close'].rolling(window=7).mean()

                # Calculate the '14 DAYS MA' column (Close fourteen days moving average)
                df['14 DAYS MA'] = df['Close'].rolling(window=14).mean()

                # Calculate the '21 DAYS MA' column (Close twenty-one days moving average)
                df['21 DAYS MA'] = df['Close'].rolling(window=21).mean()

                # Calculate the '7 DAYS STD DEV' column (Close standard deviation for the past seven days)
                df['7 DAYS STD DEV'] = df['Close'].rolling(window=7).std()

                # Save the processed DataFrame to a new CSV file in the output directory
                output_file_path = os.path.join(output_directory, file_name)
                df.to_csv(output_file_path, index=False)

                print(f"File: {file_name}, Columns and calculations added, saved to '{output_file_path}'")
            else:
                print(f"File: {file_name} is empty.")
        else:
            print(f"File: {file_name} is not a CSV file.")
else:
    print(f"'{destination_directory_15y}' does not exist.")
```

```python
import os
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

source_directory = '/Users/haojiechu/Documents/Python_documents/python_excercises/stock_data_15y_reversed3'
output_directory = '/Users/haojiechu/Documents/Python_documents/python_excercises/stock_data_15y_scaled'

# Ensure the source directory exists
if os.path.exists(source_directory):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # List all files in the source directory
    files_in_source = [file for file in os.listdir(source_directory) if os.path.isfile(os.path.join(source_directory, file))]

    # Create a MinMaxScaler
    scaler = MinMaxScaler()

    # Loop through each file
    for file_name in files_in_source:
        file_path = os.path.join(source_directory, file_name)

        # Check if the file is a CSV file (you can modify the condition for other file types)
        if file_name.endswith('.csv'):
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)

            # Check if the DataFrame has at least one row
            if not df.empty:
                # Exclude the 'Date' column from scaling
                columns_to_scale = [col for col in df.columns if col != 'Date']

                # Scale the specified columns using Min-Max scaling
                df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

                # Save the scaled DataFrame to a new CSV file in the output directory
                output_file_path = os.path.join(output_directory, file_name)
                df.to_csv(output_file_path, index=False)

                print(f"File: {file_name}, Data scaled, saved to '{output_file_path}'")
            else:
                print(f"File: {file_name} is empty.")
        else:
            print(f"File: {file_name} is not a CSV file.")
else:
    print(f"'{source_directory}' does not exist.")
```
